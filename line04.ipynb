```python
HTTP 데이터 저장 서버
30점
  문제의 입력값은 각 언어의 표준입력(stdin) 함수를, 출력값은 표준출력(stdout) 함수를 사용해주세요.
HTTP 데이터 저장 서버
코니는 사용자별로 데이터를 저장할 수 있는 서비스를 제공하고자 한다. 스펙이 아래와 같이 주어졌을 때, 코니를 도와 서버를 구현하시오.

HTTP Request를 받고, 처리하는 API는 다음과 같이 3개의 Endpoint가 있다.
Endpoint	Method	Parameters	Response	Description
/users/{UserId}	POST			사용자를 생성한다.
/users/{UserId}/data	POST	data		사용자 데이터를 저장한다.
/users/{UserId}/data	GET		data	User ID로 지정된 사용자의 Data를 리턴한다.
사용자 생성 API
URL을 통해 전달받은 UserId로 사용자를 생성하는 API이다.
/users/{UserId}를 Endpoint로 가지며 POST Method를 이용한다.
POST /users/CONY
성공적으로 생성했다면 201 코드와 그에 대응되는 메시지를 리턴한다.
이미 존재하는 UserId라면 403 코드와 그에 대응되는 메시지를 리턴한다.
GET Method를 이용하여 Request를 보내는 경우, 405 코드와 그에 대응되는 메시지를 리턴한다.
201 CREATED
403 FORBIDDEN
405 METHOD_NOT_ALLOWED
사용자 데이터 저장 API
UserId를 갖는 사용자의 데이터를 저장하는 API이다.
/users/{UserId}/data Endpoint로 가지며 POST Method를 통하여 사용자 데이터를 저장한다.
Endpoint에 UserId가 명시되어 전달된다.
저장할 Data는 Message Body를 통하여 전달된다.
value=DATA 형식으로 데이터가 전달된다. 해당 형식을 위반하는 request는 존재하지 않는다.
Data는 공백과 =이 없는 String이다.
동일 사용자에게 API가 연속적으로 불리면 Overwrite되어 가장 최근 Data만 저장된다.
POST /users/CONY/data value=TEST_DATA
성공하면 200 코드와 그에 대응되는 메시지를 리턴한다.
UserId를 갖는 사용자가 없는 경우, 403 코드와 그에 대응되는 메시지를 리턴한다.
GET Method를 이용하여 Request를 보내는 경우, 405 코드와 그에 대응되는 메시지를 리턴한다.
200 OK
403 FORBIDDEN
405 METHOD_NOT_ALLOWED
사용자 데이터 조회 API
UserId를 갖는 사용자의 저장된 데이터를 조회하는 API이다.
사용자 데이터 저장 API와 동일한 /users/{UserId}/data Endpoint를 가지며 GET Method를 통하여 사용자 데이터를 조회한다.
GET /users/CONY/data
전달된 UserId에 해당하는 사용자의 Data를 200 코드 및 메시지와 함께 Message Body에 리턴한다.
UserId를 갖는 사용자가 없는 경우, 403 코드와 그에 대응되는 메시지를 리턴한다.
만약 UserId를 갖는 사용자는 존재하지만, 저장된 데이터가 없다면 404 코드 및 메시지를 리턴한다.
POST Method를 이용하여 Request를 보내는 경우, 405 코드와 그에 대응되는 메시지를 리턴한다.
200 OK TEST_DATA
403 FORBIDDEN
404 NOT_FOUND
405 METHOD_NOT_ALLOWED
입력형식
표준 입력을 통하여 입력받는다.
HTTP Request의 개수를 입력받는다.
그다음 줄부터는 입력받은 개수만큼 HTTP Request가 입력된다.
Request는 다음 형식을 따른다.
주어지는 입력에서는 지정된 Endpoint에서 요구하는 Parameter가 누락되는 케이스는 존재하지 않는다.
[HTTP Method 명] [Endpoint] [전달할 Data가 필요한 경우 Message Body]
출력 형식
각 Request 처리결과에 대한 Response를 출력한다.
Response는 다음 형식을 따른다.
[Response Code] [Response Message] [결과 리턴이 필요한 경우 Message Body]
상황별 Response Code 요약
상황별로 출력해야 하는 Response Code와 매칭되는 Response Message를 정리하면 다음과 같다.
Code	Message	Description
200	OK	사용자 데이터 저장 API에서 사용자 Data를 성공적으로 저장한 경우 
사용자 데이터 조회 API에서 사용자 Data를 성공적으로 리턴한 경우
201	CREATED	사용자 생성 API에서 사용자를 성공적으로 생성한 경우
403	FORBIDDEN	사용자 생성 API에서 사용자 생성 시 이미 존재하는 UserID인 경우 
존재하지 않는 사용자를 대상으로 데이터 저장 혹은 조회 API를 호출한 경우
404	NOT_FOUND	존재하지 않는 API를 호출한 경우 
사용자 데이터 조회 API 호출 시 해당 사용자에게 저장된 Data가 존재하지 않는 경우
405	METHOD_NOT_ALLOWED	지정된 Endpoint에서 지원하지 않는 Method를 이용하여 호출한 경우 (예: 사용자 생성 API를 GET으로 호출한 경우)
입력에서 여러 예외 케이스가 중복되는 경우, 다음과 같은 우선순위를 따른다.
Priority	ERROR
1 (가장 높음)	NOT_FOUND (존재하지 않는 API의 경우)
2	METHOD_NOT_ALLOWED
3	FORBIDDEN
4 (가장 낮음)	NOT_FOUND (사용자의 데이터가 존재하지 않는 경우)

입/출력 예시
⋇ 입출력 형식을 잘 지켜주세요.
␣ : 공백
↵ : 줄바꿈
−⇥ : 탭
보기 입력 1
8
POST /users/CONY
POST /users/BROWN
POST /users/CONY/data value=WHERE_IS_SALLY
GET /users/CONY/data
GET /users/BROWN/data
GET /users/EDWARD/data
GET /abc
GET /users/CONY
출력 1
201 CREATED
201 CREATED
200 OK
200 OK WHERE_IS_SALLY
404 NOT_FOUND
403 FORBIDDEN
404 NOT_FOUND
405 METHOD_NOT_ALLOWED
```
