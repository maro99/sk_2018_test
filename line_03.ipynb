강 탐사
여러 명의 탐험가가 강의 물길을 탐사하기 위해 힘을 모았다. 탐험가는 각자 자기가 원하는 지천의 시작점 또는 합류점에서 출발하여 뗏목을 타고 하류로 이동하며 경로를 노트에 기록한다. 각 탐험가들은 자기가 원하는 합류점에서 탐사를 마치고 노트를 본부에 제출한다.

다음은 각 탐험가들이 제출한 노트와 그 노트를 이용해 그린 강의 물길이다.

# 하류 방향
탐험가들이 제출한 탐사 노트를 분석하여 강의 하구(mouth)들을 출력하는 프로그램을 작성하시오.

제약 조건
탐험가는 강의 중간 합류점에서 출발할 수도 있다.
탐험가는 하류 방향으로만 이동할 수 있다.
탐험가가 지나가다가 빠뜨리는 합류점은 없다.
시작점과 합류점을 나타내는 X의 범위는 1<= X <= 1000 이다.
상류의 여러 지천들이 하류에서 합쳐질 수는 있지만 다른 여러 지천으로 나누어지는 경우는 없다. 이 경우는 탐험가가 노트를 잘못 입력하였으므로
-1을 출력한다.

노트는 시작점 또는 합류점의 기호(자연수)들을 공백으로 구분하여 순서대로 입력한다.

```python
# 일단 내생각에는 받은 입력으로 
# 모두의 알고리즘에서 그래프 문제 푼것 처럼 각 point에 연결된 point를 키 벨류 형태로 어떤 그룹에 입력을 해 두자. 
# 그리고 이것을 완전 탐색 하면서 경로를 쌓아가는데 
# 한가지 조건은 앞서서 저장된 값보다 뒤에 저장되는 값이 더 커야 한다는 거다. 

# 그런데 첫번째 예와 같이두개의 하구가 있는 경우에는 어떻게 해야 될지 모르겠다. 

def get_point_group_dict(lista):

    
    point_group = dict()
    
    for row in lista:
        row_list = row.split(" ") # ['1', '2', '4', '7'] 
        
        if len(row_list)>1: #2 개이상의 원소 가진다면 
            for index, item in enumerate(row_list): # 해당 인댁스의 앞 뒤를 자신을 key로하는 value값에 추가한다. 
                
                
                
                
                if index == 0: # 맨 앞이라면 그다음 원소만 dict에 넣는다.
                    # 만약 이 item에 해당하는 dict의 key, value가 생성된적 없다면 저장해 주고 루프 넘어간다. 
                    if not point_group.get(item):
                        point_group[item] = [row_list[index+1]]
                    else:
                        if row_list[index+1] not in point_group[item]:
                            point_group[item].append(row_list[index+1])
                    
                    
                elif index == len(row_list) -1: # 맨 뒤라면 그 앞으 원소만 dict에 넣는다. 
                    # 만약 이 item에 해당하는 dict의 key, value가 생성된적 없다면 저장해 주고 루프 넘어간다. 
                    if not point_group.get(item):
                        point_group[item] = [row_list[index-1]]
                    else:
                        if row_list[index-1] not in point_group[item]:
                            point_group[item].append(row_list[index-1])
                            
                            
                else: # 만약 0<x<n-1 사이 값이면 앞뒤를 dict에 넣는다. 
                    # 만약 이 item에 해당하는 dict의 key, value가 생성된적 없다면 저장해 주고 루프 넘어간다. 
                    if not point_group.get(item):
                        point_group[item] = [row_list[index+1],row_list[index-1]]
                    else:
                        if row_list[index+1] not in point_group[item]:
                            point_group[item].append(row_list[index+1])
                            
                        if row_list[index-1] not in point_group[item]:
                            point_group[item].append(row_list[index-1])
            
            
            
            
        else: # 1개 이하라면 그냥 지나간다.
            pass
        
        
    
    return point_group




def get_hagu(point_group_dict):
    # 모든 포인트들에 대해서 정방향으로 (작은--> 큰)이동하고
    # 더이상 연결점 없을때 까지 경로를 가본뒤 출력해본다. 
    
    result_list = []
    for start in point_group_dict:
        
        print(f'{start}@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')

#         start ='1'


        waited_list = [] # 방문할 리스트. 역순이어서는 안된다. 
        done_list = [] # 방문할 리스트에 추가가 완료된 것들 담은 리스트.

        waited_list.append(start)
        done_list.append(start)

        while(waited_list):
            until_path = waited_list.pop(0) # 지금까지 왔던 경로 여기 저장되 있다.
            this_point_key = until_path.split(" ")[-1] # dict에서 참조할 주최가 될 point

            print(f'this_point_key:  {this_point_key}:')
            print(f'point_group_dict[this_point_key]: {point_group_dict[this_point_key]}')
            for value in point_group_dict[this_point_key] :# 이 this_point와 인접한 point들 돌면서 체크후 리스트 추가해좀.
                print(f'value: {value}')
                print(f'done_list: {done_list}')
                if value not in done_list and int(value) > int(this_point_key):
                    waited_list.append(until_path+" "+ value)
                    done_list.append(value)

            print(waited_list)


            print('################################################################')
            
            
            
        # while문 나온후 done list 마지막에 있는것을 result_list에 추가해 준다. 
        
        if done_list[-1] not in result_list:
            result_list.append(done_list[-1])
        
        
        print(f'{start}@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
    
    return result_list
```

```python
lista = ["1 2 4 7", "3 4 7 9", "5 8 9","6 8","10 12 14", "11 12 14 17", "15 17", "13 16 17"]
listb = ["8", "3 8 11 12", "2 7 11", "1 6 7","10 12", "3 8","4 8 11", "5 9 11 12","3 8 11"]


point_group_dict = get_point_group_dict(lista)
# print(point_group_dict)

result = get_hagu(point_group_dict)
print(result)
```
